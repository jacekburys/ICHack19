```python
def loadImage(path='images/kanye.jpg'):
    global img
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
loadImage()
plt.imshow(img)
plt.show()
```

```python
def loadOrange(path='images/orange.png'):
    global orange_img
    orange = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    orange = cv2.cvtColor(orange, cv2.COLOR_BGRA2RGBA)
    orange_img = orange
    return orange

loadOrange()
plt.imshow(orange_img)
plt.show()
```

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

def getLandmarks(im):
    rects = detector(im, 1)
    
    if len(rects) > 1:
        pass
        #raise TooManyFaces
    if len(rects) == 0:
        pass
        #raise NoFaces

    h, w, c = im.shape
    #border = [(0,0), (0, h), (w,h), (w,0), (w//2, 0)]
    border = [(0,0), (0, h-1), (w-1,h-1), (w-1,0), (w//2, 0)]

        
    return [[(p.x, p.y) for p in predictor(im, rect).parts()] + border for rect in rects]
```

```python
loadImage()
landmarks = getLandmarks(img)
print(len(landmarks))
print(landmarks)
font = cv2.FONT_HERSHEY_SIMPLEX
for i,landmark in enumerate(landmarks[0]):
    cv2.circle(img, landmark, 3, 255)
    cv2.putText(img, str(i), landmark, font, 0.5, 255)

plt.imshow(img)
plt.show()
```

```python
def findFaceCentre(landmarks):
    #print(landmarks)
    n = len(landmarks)
    sx = 0
    sy = 0
    for landmark in landmarks:
        sx += landmark[0]
        sy += landmark[1]
    return (sx//n, sy//n)
```

```python
# returns (width, height)
def findFaceDims(landmarks):
    def lineLength(p1, p2):
        x = abs(p1[0]-p2[0])
        y = abs(p1[1]-p2[1])
        return math.sqrt(x*x + y*y)
    a = landmarks[0]
    b = landmarks[16]
    
    c = landmarks[8]
    d = landmarks[27]
    
    return (lineLength(a, b), lineLength(c, d))
```
